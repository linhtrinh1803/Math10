Refer to the [gallery](https://seaborn.pydata.org/examples/index.html) or [other tutorials](https://seaborn.pydata.org/tutorial.html) on the website, and generate more fancy figures with this dataset!

```python
fig4 = sns.displot(data=iris, x="sepal_width", hue="species", col="species")
fig4.set(xlabel = 'SepalWidthCm')
```

```python
fig5 = sns.jointplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
```

```python
sns.jointplot(data=iris, x="petal_length", y="petal_width", hue="species", kind="hist")
```

```python
fig6 = sns.lmplot(data=iris, x="sepal_length", y="sepal_width", hue="species", height=5)
fig6.set(xlabel = 'SepalLengthCm', ylabel = 'SepalWidthCm')
```

```python
fig7 = sns.lineplot(x="petal_length", y="petal_width",hue="species", data=iris)
fig7.set(xlabel='PetalLengthCm', ylabel='PetalWidthCm')
```
